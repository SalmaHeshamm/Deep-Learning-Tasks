```python
import numpy as np
```

```python
x = [1,2,3,4]

input_w = [0.5, 0.6, 0.7, 0.8]
h_w = [0.1, 0.2, 0.3, 0.4, 4]
bias = [0, 0, 0, 0, 0]

h = 0
c = 0
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))


def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
for i in range(len(x)):

    c = (sigmoid(((input_w[0] * x[i]) + (h_w[0] * h) + bias[0])) * c ) +\
        (sigmoid(((input_w[1] * x[i]) + (h_w[1] * h) + bias[1])) * tanh(((input_w[2] * x[i]) + (h_w[2] * h) + bias[2])))

    h = ( sigmoid(((input_w[3] * x[i]) + (h_w[3] * h) + bias[3])) * tanh(c))

    y = (h_w[4] * h ) + bias[4]

    print(f'Time step {x[i]}')
    print(f' Cell state: {c}')
    print(f' Hidden state: {h}')
    print()

print(f' Predicted next value for t = {i+1} is: {y}')
```
